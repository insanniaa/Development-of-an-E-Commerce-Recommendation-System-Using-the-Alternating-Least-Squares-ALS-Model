# **Pendahuluan**

Sistem rekomendasi merupakan salah satu elemen utama dalam meningkatkan pengalaman pengguna pada platform e-commerce. Dengan memanfaatkan data interaksi pelanggan, sistem ini dapat memberikan rekomendasi produk yang relevan dan personal. Proyek ini berfokus pada pembangunan sistem rekomendasi berbasis Machine Learning, menggunakan teknik Alternating Least Squares (ALS) untuk menghasilkan rekomendasi yang disesuaikan dengan preferensi pelanggan. Alternating Least Squares (ALS) adalah teknik matrix factorization yang dirancang untuk mengelola data yang jarang (sparse data), yang sering ditemukan dalam sistem rekomendasi. Menurut He et al. (2017), ALS sangat cocok untuk menangani data umpan balik implisit (implicit feedback), seperti klik, riwayat pembelian, atau waktu interaksi, yang umum dalam sistem rekomendasi e-commerce. Selain itu, ALS mampu memproses data skala besar secara efisien karena algoritmanya membagi optimasi menjadi langkah-langkah terpisah yang dapat diproses secara paralel, menjadikannya pilihan ideal untuk pengolahan data yang kompleks (Chen et al., 2017). Model ALS dipilih karena kehandalannya dalam memberikan hasil yang akurat, bahkan pada dataset dengan banyak nilai kosong (sparse). Sebagaimana dinyatakan oleh Liang et al. (2018), pendekatan berbasis ALS sering digunakan dalam sistem rekomendasi berbasis collaborative filtering karena kemampuannya untuk menangkap pola preferensi pelanggan secara efektif, terutama dalam skenario dengan data yang tidak lengkap.

Dalam proyek ini, data transaksi pelanggan, seperti riwayat pembelian dan rating produk, digunakan untuk membangun model yang dapat mengidentifikasi pola preferensi pelanggan dan memberikan rekomendasi produk secara proaktif. Pendekatan ini tidak hanya memberikan pengalaman belanja yang lebih personal bagi pelanggan, tetapi juga nilai strategis bagi bisnis. Dengan rekomendasi yang akurat, perusahaan dapat meningkatkan penjualan, mengoptimalkan pengelolaan inventaris, dan membangun loyalitas pelanggan. Sebagaimana disampaikan oleh Li et al. (2020)  penerapan sistem rekomendasi yang berbasis deep learning terbukti dapat meningkatkan tingkat konversi penjualan hingga 30% dibandingkan dengan pendekatan rekomendasi konvensional. Proyek ini mencakup beberapa tahapan penting, mulai dari eksplorasi data, pembangunan model rekomendasi, hingga evaluasi kinerja model dengan menggunakan metrik Root Mean Squared Error (RMSE) dan Mean Absolute Error (MAE). Hasil dari proyek ini diharapkan dapat memberikan wawasan berharga bagi perusahaan dalam meningkatkan efisiensi operasional sekaligus memaksimalkan kepuasan pelanggan.

Menangani missing values pada kolom product_weight_g, product_length_cm, product_height_cm, dan product_width_cm dengan mengganti baris yang mengandung missing value menjadi nilai median dari masing-masing kolom tersebut.

```python
#menghitung median untuk setiap kolom
median_weight = products_df.approxQuantile('product_weight_g', [0.5], 0)[0]
median_length = products_df.approxQuantile('product_length_cm', [0.5], 0)[0]
median_height = products_df.approxQuantile('product_height_cm', [0.5], 0)[0]
median_width = products_df.approxQuantile('product_width_cm', [0.5], 0)[0]

#mengganti missing value dengan median yang telah dihitung
products_df = products_df.withColumn('product_weight_g',
                                     F.coalesce(F.col('product_weight_g'), F.lit(median_weight)))
products_df = products_df.withColumn('product_length_cm',
                                     F.coalesce(F.col('product_length_cm'), F.lit(median_length)))
products_df = products_df.withColumn('product_height_cm',
                                     F.coalesce(F.col('product_height_cm'), F.lit(median_height)))
products_df = products_df.withColumn('product_width_cm',
                                     F.coalesce(F.col('product_width_cm'), F.lit(median_width)))
```

Memeriksa kembali apakah missing value pada ketiga kolom tersebut berhasil ditangani.

```python
null_check = products_df.select(
    [F.count(F.when(F.col(col).isNull(), 1)).alias(col) for col in ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']]
)
null_check.show()
```

Berdasarkan informasi di atas, dapat diketahui bahwa missing value pada kolom product_weight_g, product_length_cm, product_height_cm, dan product_width_cm berhasil ditangani. Dengan demikian, semua missing values pada kolom products_df berhasil ditangani.

**RVM Analysis**

## **Kapan terakhir kali pelanggan melakukan transaksi?**

```python
#mengelompokkan data berdasarkan customer_id dan menghitung nilai RFM
rfm_df = final_df.groupBy("customer_id").agg(
    F.max("order_approved_at").alias("max_order_timestamp"),
    F.countDistinct("order_id").alias("frequency"),
    F.sum("price").alias("monetary")
)

#mendapatkan tanggal terbaru sebagai referensi untuk menghitung recency
recent_date = final_df.agg(F.max("order_approved_at")).collect()[0][0]

#menghitung nilai recency dalam hari
rfm_df = rfm_df.withColumn(
    "recency", F.datediff(F.lit(recent_date), F.col("max_order_timestamp"))
)

rfm_df = rfm_df.drop("max_order_timestamp")
rfm_df.sort("recency", ascending=True).show()
```

Jika hari ini adalah Hari ke-0, maka transaksi terakhir pelanggan dapat dihitung dengan mengurangi nilai pada kolom "recency" dari hari ini. Sebagai contoh, pelanggan dengan customer_id 28336 memiliki nilai recency 0, yang berarti pelanggan ini melakukan transaksi terakhir pada hari yang sama dengan hari ini. ### Analisis Musiman dan Tren Jangka Panjang

```python
#Analisis Musiman (Seasonality) dan Tren Jangka Panjang
from pyspark.sql.functions import month, year, dayofweek, col, sum, count

#menambahkan kolom bulan, tahun, dan hari dalam minggu untuk analisis musiman
final_df = final_df.withColumn("year", year("order_purchase_timestamp"))
final_df = final_df.withColumn("month", month("order_purchase_timestamp"))
final_df = final_df.withColumn("day_of_week", dayofweek("order_purchase_timestamp"))

#total penjualan per bulan
monthly_sales = final_df.groupBy("year", "month").agg(
    sum("price").alias("total_sales"),
    count("order_id").alias("total_orders")
).orderBy("year", "month")

monthly_sales.show()

monthly_sales_pd = monthly_sales.toPandas()

monthly_sales_pd['year_month'] = monthly_sales_pd['year'].astype(str) + '-' + monthly_sales_pd['month'].astype(str)

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales_pd['year_month'], monthly_sales_pd['total_sales'], marker='o')
plt.xticks(rotation=90)
plt.title("Total Penjualan Per Bulan")
plt.xlabel("Bulan-Tahun")
plt.ylabel("Total Penjualan")
plt.tight_layout()
plt.show()
```

Berdasarkan data di atas, dapat dilihat tren penjualan dan jumlah pesanan yang terjadi dari tahun 2016 hingga 2018. Pada awal periode, terutama pada bulan-bulan seperti September 2016 dan Desember 2016, total penjualan dan jumlah pesanan relatif rendah, dengan penjualan hanya mencapai sekitar 404.91 USD dan 10 pesanan pada bulan September 2016, serta 10.9 USD dan 1 pesanan pada bulan Desember 2016. Namun, pada tahun 2017, terjadi lonjakan signifikan dalam penjualan dan jumlah pesanan, dimulai dari Januari hingga Desember, dengan bulan Oktober 2017 dan November 2017 mencatatkan penjualan tertinggi di atas 1 juta USD dan jumlah pesanan mencapai lebih dari 12.000.

Pada tahun 2018, meskipun terdapat sedikit fluktuasi, total penjualan tetap menunjukkan angka yang tinggi, dengan bulan-bulan seperti Januari hingga Mei 2018 mencatatkan penjualan yang stabil di atas 1 juta USD setiap bulannya. Secara keseluruhan, data ini menunjukkan adanya tren positif yang berkelanjutan dalam penjualan dan jumlah pesanan, dengan puncak yang terlihat pada bulan Oktober dan November 2017. ## Feature selection

Kolom yang relevan untuk analisis dipilih dari dataset final_df, yaitu customer_id, product_id, dan review_score. Kolom customer_id merepresentasikan identitas pelanggan, product_id adalah identitas produk yang dinilai, dan review_score adalah nilai atau rating yang diberikan oleh pelanggan terhadap produk tersebut. Setelah itu, fungsi dropna() digunakan untuk menghapus baris yang mengandung nilai kosong (null) di salah satu dari ketiga kolom tersebut. Ini penting agar data yang digunakan untuk melatih model tidak memiliki nilai yang hilang yang dapat mengganggu proses pelatihan dan prediksi model.

```python
from pyspark.ml.recommendation import ALS
from pyspark.sql import functions as F
from pyspark.ml.evaluation import RegressionEvaluator

data = final_df.select('customer_id', 'product_id', 'review_score').dropna()
```

## Merubah tipe data

Dilakukan proses konversi nilai kategorikal menjadi nilai numerik yang dapat digunakan oleh model pembelajaran mesin. Fungsi StringIndexer dari PySpark digunakan untuk mengonversi kolom customer_id dan product_id, yang awalnya berupa data kategori (identitas pelanggan dan produk), menjadi nilai numerik. Kolom customer_id diubah menjadi kolom baru customer_index, dan kolom product_id diubah menjadi kolom baru product_index. Fungsi fit() digunakan untuk melatih StringIndexer berdasarkan data yang ada, sedangkan fungsi transform() menerapkan perubahan tersebut ke dalam data.

Setelah itu, dataset yang sudah memiliki indeks numerik pada kolom customer_index dan product_index dipilih untuk digunakan dalam model rekomendasi. Kolom yang dipilih untuk model ALS adalah kolom indeks numerik customer_index, product_index, dan nilai rating review_score.

```python
from pyspark.ml.feature import StringIndexer

customer_indexer = StringIndexer(inputCol='customer_id', outputCol='customer_index')
product_indexer = StringIndexer(inputCol='product_id', outputCol='product_index')

data = customer_indexer.fit(data).transform(data)
data = product_indexer.fit(data).transform(data)

#memilihkolom indeks numerik untuk ALS
data = data.select('customer_index', 'product_index', 'review_score')
```

## Data spliting

Dataset data dibagi menjadi dua bagian: data pelatihan (training) dan data pengujian (testing) menggunakan metode randomSplit(). Fungsi ini membagi dataset secara acak berdasarkan proporsi yang ditentukan, dalam hal ini 80% data digunakan untuk pelatihan dan 20% untuk pengujian.

```python
train, test = data.randomSplit([0.8, 0.2], seed=42)
```

## Membangun model ALS

Model rekomendasi berbasis Alternating Least Squares (ALS) dibangunmenggunakan parameter yang telah ditentukan. Model ALS digunakan untuk membangun sistem rekomendasi dengan memprediksi rating yang mungkin diberikan oleh pengguna pada produk tertentu.


```python
als = ALS(
    maxIter=10,
    regParam=0.1,
    userCol='customer_index',
    itemCol='product_index',
    ratingCol='review_score',
    coldStartStrategy='drop',
    seed=42
)
```

Beberapa parameter yang diatur dalam kode ini adalah:

- maxIter=10: Menentukan jumlah iterasi maksimum yang akan dilakukan dalam proses pelatihan model. Semakin banyak iterasi, semakin baik model dapat menyempurnakan prediksi, tetapi juga membutuhkan waktu yang lebih lama untuk dilatih.
- regParam=0.1: Merupakan parameter regularisasi yang digunakan untuk menghindari overfitting. Nilai ini mengontrol seberapa besar pengaruh regularisasi terhadap model.
- userCol='customer_index': Menyatakan bahwa kolom yang merepresentasikan pengguna (customer) adalah customer_index yang sudah diubah menjadi indeks numerik.
- itemCol='product_index': Menyatakan bahwa kolom yang merepresentasikan produk adalah product_index, yang juga telah diubah menjadi indeks numerik.
- ratingCol='review_score': Menyatakan bahwa kolom yang digunakan sebagai rating atau nilai yang diberikan oleh pengguna terhadap produk adalah review_score.
- coldStartStrategy='drop': Menentukan strategi untuk menangani kasus di mana ada data yang tidak tersedia (misalnya, prediksi untuk pengguna atau produk baru yang tidak memiliki rating). Dengan strategi ini, baris-baris tersebut akan dihapus agar model tidak gagal saat memprediksi nilai kosong.
- seed=42: Menetapkan nilai seed untuk memastikan bahwa proses pelatihan yang acak menghasilkan hasil yang dapat direproduksi.

Dengan parameter-parameter ini, model ALS akan dilatih untuk memprediksi rating produk berdasarkan data yang ada.

## Melatih model

Model ALS yang telah dibangun sebelumnya diterapkan pada data pelatihan (train) menggunakan metode fit(). Fungsi fit() ini melatih model menggunakan data yang telah dibagi dan dipersiapkan pada tahap sebelumnya. Proses pelatihan ini bertujuan untuk menemukan pola atau hubungan antara pelanggan, produk, dan rating yang diberikan oleh pelanggan. Model ALS akan mempelajari matriks faktor tersembunyi yang menggambarkan preferensi pengguna dan karakteristik produk, sehingga dapat memprediksi rating yang mungkin diberikan oleh pelanggan terhadap produk yang belum mereka beri penilaian.

```python
als_model = als.fit(train)
```

Hasil dari proses ini adalah model yang terlatih, yang disimpan dalam variabel als_model dan dapat digunakan untuk melakukan prediksi pada data pengujian.

## Menguji model

Model ALS yang telah dilatih (als_model) digunakan untuk memprediksi rating pada data pengujian (test) dengan metode transform(). Fungsi transform() akan menerapkan model yang telah dilatih pada data pengujian untuk menghasilkan prediksi nilai rating yang diberikan oleh pelanggan terhadap produk yang belum dinilai sebelumnya.

```python
predictions = als_model.transform(test)
```

Hasil prediksi ini akan berupa sebuah dataset yang berisi kolom-kolom seperti customer_index, product_index, review_score yang diprediksi, serta informasi lainnya. Dataset prediksi ini akan digunakan untuk mengevaluasi kinerja model.

## Mengevaluasi model

Dua metrik evaluasi yang digunakan untuk mengukur kinerja model rekomendasi yang telah dibangun adalah RMSE (Root Mean Square Error) dan MAE (Mean Absolute Error).

RMSE dihitung untuk mengukur seberapa besar perbedaan antara nilai yang diprediksi oleh model dan nilai yang sebenarnya (rating yang diberikan oleh pelanggan) dan MAE mengukur rata-rata kesalahan absolut antara nilai yang diprediksi dan nilai yang sebenarnya tanpa memberikan bobot lebih pada kesalahan besar.

```python
#mengevaluasi menggunakan RMSE (Root Mean Square Error)
evaluator_rmse = RegressionEvaluator(metricName='rmse', labelCol='review_score', predictionCol='prediction')
rmse = evaluator_rmse.evaluate(predictions)
print(f"Root Mean Square Error (RMSE): {rmse}")

#mengevaluasi menggunakan MAE (Mean Absolute Error)
evaluator_mae = RegressionEvaluator(metricName='mae', labelCol='review_score', predictionCol='prediction')
mae = evaluator_mae.evaluate(predictions)
print(f"Mean Absolute Error (MAE): {mae}")
```

Nilai RMSE yang diperoleh adalah sekitar 0.55, yang menunjukkan bahwa secara rata-rata, prediksi rating model memiliki kesalahan sekitar 0.55 poin dari nilai sebenarnya. Sementara itu, Nilai MAE yang diperoleh adalah sekitar 0.14, yang berarti rata-rata kesalahan prediksi model sekitar 0.14 poin. MAE memberikan gambaran yang lebih sederhana tentang seberapa akurat model dalam memprediksi nilai tanpa dipengaruhi oleh kesalahan besar.

Maka, dapat disimpulkan bahwa model rekomendasi yang dibangun memiliki performa yang cukup baik dengan kesalahan prediksi yang relatif rendah. Meskipun RMSE menunjukkan kesalahan yang sedikit lebih besar (sekitar 0.55), ini menunjukkan bahwa model masih mampu memberikan prediksi rating yang cukup akurat secara keseluruhan. Namun, nilai MAE yang lebih rendah (sekitar 0.14) menunjukkan bahwa mayoritas prediksi model berada dekat dengan nilai sebenarnya, tanpa dipengaruhi oleh prediksi yang sangat menyimpang.

## Membuat rekomendasi

### Rekomendasi untuk semua pelanggan

Model ALS yang telah dibangun digunakan untuk memberikan rekomendasi produk kepada semua pelanggan dengan metode recommendForAllUsers(numItems=5). Fungsi ini akan menghasilkan daftar 5 produk teratas yang diprediksi akan disukai oleh setiap pelanggan berdasarkan preferensi dan pola pembelian mereka yang telah dipelajari oleh model.

```python
#mmberikan rekomendasi produk untuk semua pelanggan
customer_recommendations = als_model.recommendForAllUsers(numItems=5)
customer_recommendations.show(truncate=False)
```

Mendekode dan menampilkan rekomendasi produk yang telah diberikan oleh model ALS dalam bentuk yang lebih mudah dipahami.

```python
from pyspark.ml.feature import StringIndexer

customer_indexer = StringIndexer(inputCol="customer_id", outputCol="customer_index")
customer_model = customer_indexer.fit(final_df)

customer_mapping = customer_model.labels
customer_index_mapping = spark.createDataFrame(
    [(i, customer_mapping[i]) for i in range(len(customer_mapping))],
    ["customer_index", "customer_id"]
)
customer_index_mapping.show()
```

```python
product_indexer = StringIndexer(inputCol="product_id", outputCol="product_index")
product_model = product_indexer.fit(final_df)

product_mapping = product_model.labels
product_index_mapping = spark.createDataFrame(
    [(i, product_mapping[i]) for i in range(len(product_mapping))],
    ["product_index", "product_id"]
)
product_index_mapping.show()
```

Menggabungkan hasil rekomendasi dengan data pemetaan pelanggan (customer_index_mapping) menggunakan join berdasarkan kolom customer_index

```python
decoded_recommendations = customer_recommendations.join(customer_index_mapping, "customer_index", "left")
decoded_recommendations.show()
```

Menggunakan fungsi explode, daftar rekomendasi produk untuk setiap pelanggan diubah menjadi baris terpisah, sehingga setiap baris berisi satu rekomendasi produk untuk seorang pelanggan. Kolom recommendation yang sebelumnya berisi array dipecah, dan informasi produk yang direkomendasikan diambil dengan menggunakan getItem untuk memperoleh product_index dan score (nilai rating).

Kemudian, data ini digabungkan lagi dengan pemetaan produk (product_index_mapping) untuk mengonversi product_index menjadi product_id, menghasilkan dataset akhir yang berisi customer_id, product_id, dan score untuk setiap rekomendasi.

```python
from pyspark.sql.functions import explode, col

recommendations_exploded = decoded_recommendations.select(
    "customer_index", "customer_id", explode("recommendations").alias("recommendation")
)

recommendations_exploded = recommendations_exploded.withColumn(
    "product_index", col("recommendation").getItem("product_index")
).withColumn(
    "score", col("recommendation").getItem("rating")
).drop("recommendation")

final_recommendations = recommendations_exploded.join(product_index_mapping, "product_index", "left")
final_recommendations.select("customer_id", "product_id", "score").show()
```

Hasil akhirnya adalah daftar rekomendasi produk yang dipersonalisasi untuk setiap pelanggan, yang dapat digunakan untuk menampilkan saran produk yang lebih relevan dan personal kepada pengguna di platform e-commerce. Data ini bertujuan untuk meningkatkan pengalaman pengguna dan potensi konversi penjualan di platform e-commerce.

### Rekomendasi untuk produk tertentu

 Model ALS yang telah dibangun digunakan untuk memberikan rekomendasi pelanggan untuk produk tertentu dengan metode recommendForAllItems(numUsers=5). Fungsi ini akan menghasilkan daftar 5 pelanggan teratas yang diprediksi akan menyukai setiap produk berdasarkan pola pembelian dan preferensi yang telah dipelajari oleh model.

```python
#memberikan rekomendasi pelanggan untuk produk tertentu
product_recommendations = als_model.recommendForAllItems(numUsers=5)
product_recommendations.show(truncate=False)
```

Fungsi decode_recommendations_udf digunakan untuk mendekodekan rekomendasi produk yang dihasilkan oleh model ALS. Fungsi ini menerima dua parameter: rec_list, yang berisi daftar rekomendasi untuk setiap produk, dan mapping_dict, yang merupakan kamus yang memetakan product_index ke product_id yang sebenarnya. Fungsi ini akan memproses setiap rekomendasi dalam rec_list, mengambil nilai product_index dan rating, kemudian mencari product_id yang sesuai dari mapping_dict. Jika product_index ditemukan dalam kamus, maka rekomendasi tersebut akan dimasukkan ke dalam daftar hasil dengan format yang berisi product_id dan rating.

Selanjutnya, decode_udf mendefinisikan UDF (User Defined Function) yang menggunakan decode_recommendations_udf untuk mendekodekan rekomendasi produk dalam kolom recommendations. Fungsi ini kemudian diterapkan pada dataframe product_recommendations untuk menghasilkan kolom baru yang disebut decoded_recommendations. Kolom ini berisi hasil rekomendasi yang sudah terdekode, yang menampilkan product_id dan rating untuk setiap produk yang direkomendasikan.

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import ArrayType, StructType, StructField, StringType, FloatType

mapping_dict = product_index_mapping.rdd.map(lambda row: (row.product_index, row.product_id)).collectAsMap()

def decode_recommendations_udf(rec_list, mapping_dict):
    decoded_recs = []
    if rec_list:
        for rec in rec_list:
            product_index = rec[0]
            rating = rec[1]
            if product_index is not None and rating is not None:
                product_id = mapping_dict.get(product_index, "unknown")
                if product_id != "unknown":
                    decoded_recs.append({"product_id": product_id, "rating": rating})
    return decoded_recs

decode_udf = udf(lambda rec_list: decode_recommendations_udf(rec_list, mapping_dict), ArrayType(StructType([
    StructField("product_id", StringType()),
    StructField("rating", FloatType())
])))

product_recommendations = product_recommendations.withColumn("decoded_recommendations", decode_udf("recommendations"))

product_recommendations.select("product_index", "decoded_recommendations").show(truncate=False)
```

Hasil dari proses ini adalah dataset yang berisi rekomendasi untuk produk tertentu, di mana setiap product_index memiliki daftar produk yang direkomendasikan beserta rating-nya. Data ini memungkinkan kita untuk menargetkan pelanggan yang kemungkinan tertarik dengan produk-produk tertentu. Ini dapat mendukung strategi pemasaran yang lebih terfokus, di mana produk yang direkomendasikan dapat dipromosikan kepada pelanggan yang relevan, meningkatkan peluang konversi penjualan. Tampilan yang menunjukkan product_index dan rekomendasi produk yang terdekodekan ini memberi informasi yang lebih jelas tentang produk yang direkomendasikan dan rating yang diberikan oleh pelanggan, yang pada gilirannya dapat digunakan untuk memberikan pengalaman belanja yang lebih personal dan relevan bagi pelanggan di platform e-commerce.
